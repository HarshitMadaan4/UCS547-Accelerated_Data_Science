Find the most frequent value and their indice(s) in the following arrays:
i.
x = np.array([1,2,3,4,5,1,2,1,1,1])
ii.
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
import numpy as np

def most_frequent(arr):
    values, counts = np.unique(arr, return_counts=True)
    max_count = np.max(counts)
    most_freq_values = values[counts == max_count]

    result_indices = {}
    for val in most_freq_values:
        result_indices[val] = np.where(arr == val)[0].tolist()

    return most_freq_values, result_indices

x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

print(f"For x: {x}")
freq_x_values, freq_x_indices = most_frequent(x)
print(f"Most frequent value(s): {freq_x_values}, Indices: {freq_x_indices}")

print(f"\nFor y: {y}")
freq_y_values, freq_y_indices = most_frequent(y)
print(f"Most frequent value(s): {freq_y_values}, Indices: {freq_y_indices}")
```

For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
find i. Sum of all elements
ii.
iii.
Sum of all element’s row-wise
Sum of all element’s column-wise

```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

#i)sum of all elements
print(np.sum(gfg))

#ii)sum of elements row-wise
print(np.sum(gfg, axis=1))

#iii)sum of elements column-wise
print(np.sum(gfg, axis=0))
```
